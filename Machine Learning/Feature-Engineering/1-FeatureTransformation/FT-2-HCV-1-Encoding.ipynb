```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,LabelEncoder
```

```python
df = pd.read_csv("customer.csv")
# Age: Numerical (Discrete); gender, purchased: Categorical (Nominal); review, education: Categorcal (Oridnal)
# Note: The purchased is a output column (i.e., Label). Label encoding is only used for output columns.
df
```

# Ordinal Encoding

```python
ordinal_data = df[['review','education']]
# Note: The sequence in categories must be organized as per requirement. In this case review: (Poor, Average, Good) = (0,1,2) and 
# (School,UG: Undergraduate, PG: PostGraduate) = (0,1,2)
ordinal = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

ordinal_encoded_data = ordinal.fit_transform(ordinal_data)
ordinal_encoded_data
```

```python
ordinal.categories_
# 0: Poor, 1: Average, 2: Good;  0: School, UG: 1, PG: 2;
```

# One Hot Encoding

```python
data = pd.read_csv("cars.csv")
data
```

```python
data['fuel'].value_counts()
```

```python
data['owner'].value_counts()
```

```python
ohe_data = pd.DataFrame(data[['fuel','owner']])
ohe_data
```

```python
# Note one hot encoding is used for Categorical Nominal Data. You can customize as per your need.
# drop = 'First' column technique. 100, 010, 001 = 00, 10, 01
ohe = OneHotEncoder()
ohe_encoded_data = ohe.fit_transform(ohe_data)
ohe_encoded_data
```

```python
ohe_encoded_data.toarray()
```

```python

# fuel
# 0. Diesel    4402
# 1. Petrol    3631
# 2. CNG         57
# 3. LPG         38

# owner
# 4 First Owner             5289
# 5. Second Owner            2105
# 6. Third Owner              555
# 7. Fourth & Above Owner     174
# 8. Test Drive Car             5


ohe_df = pd.DataFrame(ohe_encoded_data.toarray())
ohe_df
```

## One Hot Encoding Using Pandas

```python
dummies = pd.get_dummies(data[['fuel','owner']])
dummies
```

```python
data = data + dummies
data
```

# Label Encoding

```python
# Label encoding is used for the output categorical column. In this case purchased.
label_column = df['purchased']
le = LabelEncoder()
label_encoded_column= le.fit_transform(label_column)
label_encoded_column
```

```python
le.classes_
# 0: No, 1: Yes;
```
